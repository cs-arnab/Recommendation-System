```python
# URL of the CSV file in GitHub
base_url = "https://raw.githubusercontent.com/cs-arnab/Recommendation-System/master/Data%20Set/"
movies = base_url + "tmdb_5000_movies.csv"
credits = base_url + "tmdb_5000_credits.csv"
```

```python
movies = movies.merge(credits,on='title')
print(movies.shape)
movies.head(2)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(2)
```

```python
# we need 3 column to process
# movie_id, title and tags
# tags = overview	+ genres + keywords + cast + crew

# 1. drop missing data
print(movies.isnull().sum())
movies.dropna(inplace=True)
# 2. check duplicate
print(movies.duplicated().sum())
movies.drop_duplicates(inplace=True)
```

```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

# The ast.literal_eval(text) function in Python is used to safely evaluate
# an expression or a string containing a Python literal (such as a dictionary,
# list, tuple, number, string, boolean, or None). It's safer than using eval()
# because it only evaluates literals and does not execute arbitrary code.
text = '[1, 2, 3, 4, 5]'
print(text)
print(type(text))

result = ast.literal_eval(text)
print(result)  # Output: [1, 2, 3, 4, 5]
print(type(result))  # Output: <class 'list'>
# [1, 2, 3, 4, 5] <class 'str'> -> [1, 2, 3, 4, 5] <class 'list'>
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
#find out director
import json
# Sample JSON data
json_data = movies['crew'][0]

# Load JSON data
data = json.loads(json_data)

# Convert to DataFrame
df = pd.json_normalize(data)
df.head(2)
```

```python
# print(df)
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

```python
# In the 22nd century -> In, the, 22nd, century,
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(2)
```

```python
movies.head(2)
```

```python
# Sam Worthington -> SamWorthington
# Removing spaces ensures consistency, improves search accuracy, reduces
# noise, normalizes data, and enhances efficiency in recommender systems
# and data processing.

def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies.head(2)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new['tags'] = new['tags'].apply(lambda x: " ".join(x))

for i,v in enumerate(new['tags']):
  new['tags'][i] = v.lower()

print(len(new['tags']))
new.head(2)
```

```python
# ps = PorterStemmer(): This creates an instance of the PorterStemmer class,
# which is a widely used stemming algorithm in NLP. Stemming algorithms like
# Porter's reduce words to their base or root form by removing suffixes.
# Love, Loving, Loved -> love

import nltk
from nltk.stem.porter import PorterStemmer

nltk.download('punkt')  # Download the punkt tokenizer if you haven't already
ps = PorterStemmer()
def stem(text):
    tokens = nltk.word_tokenize(text)  # Tokenize the text into words
    stemmed_tokens = [ps.stem(token) for token in tokens]  # Apply stemming to each word
    return " ".join(stemmed_tokens)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
# Text to vector techniques are methods used to convert textual data into
# numerical vectors, which can be processed by machine learning algorithms.

# Here are some common text to vector techniques:

# 1. Bag of Words (BoW):
#    - Description: Represents text as a set of unique words, ignoring grammar
#    and word order.
#    - Advantages: Simple, captures word frequencies.
#    - Disadvantages: Loses word order and context.
#    - Example Techniques: CountVectorizer in scikit-learn, BoW in Natural
#    Language Toolkit (NLTK).

# 2. Term Frequency-Inverse Document Frequency (TF-IDF):
#    - Description: Weighs words based on their frequency in a document
#    relative to their frequency in the entire corpus.
#    - Advantages: Considers word importance and document specificity.
#    - Disadvantages: Ignores word order and context.
#    - Example Techniques: TfidfVectorizer in scikit-learn.

# 3. Word Embeddings (Word2Vec, GloVe, FastText):
#    - Description: Represents words as dense vectors in a continuous
#    vector space, capturing semantic relationships.
#    - Advantages: Captures word semantics and relationships.
#    - Disadvantages: Requires large amounts of data for training.
#    - Example Techniques: Word2Vec, GloVe, FastText.

# 4. Doc2Vec (Paragraph Vector):
#    - Description: Extends Word2Vec to represent entire documents as vectors.
#    - Advantages: Captures document semantics.
#    - Disadvantages: Requires large amounts of data for training.
#    - Example Techniques: Doc2Vec in gensim.

# 5. Topic Modeling (Latent Dirichlet Allocation, LDA):
#    - Description: Identifies latent topics in a collection of documents
#    and represents documents based on topic distributions.
#    - Advantages: Captures document themes and topics.
#    - Disadvantages: Requires domain knowledge for interpretation.
#    - Example Techniques: Latent Dirichlet Allocation (LDA) in gensim,
#    scikit-learn.

# 6. Transformer-based Models (BERT, GPT-3):
#    - Description: Utilizes transformer architecture to generate contextualized
#    word embeddings or document representations.
#    - Advantages: Captures context and semantics effectively.
#    - Disadvantages: Computationally intensive, requires large pre-trained
#    models.
#    - Example Techniques: BERT, GPT-3.

# Each technique has its strengths and weaknesses, and the choice depends on
# the specific task, data characteristics, and computational resources available.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
# print(vector.shape)
print(vector)
# cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
import math

def calculate_cos_theta(angle_in_degrees):
    angle_in_radians = math.radians(angle_in_degrees)
    cos_theta = math.cos(angle_in_radians)
    return cos_theta

# Example usage
def fun(angle_degrees):
  for i in range(0,angle_degrees):
    cos_value = calculate_cos_theta(i)
    print(f"The cosine of {i} degrees is {cos_value}")

angle_degrees = 91
fun(angle_degrees)
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
